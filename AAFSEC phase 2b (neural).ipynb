```python
# =============================================================================
# DAVIS - Phase 2b: Neural Model Comparison
# Goal: Compare neural models under identical LOOCV conditions
# Models: MLP-1Layer, MLP-3Layer, 1D CNN, CNN-LSTM with Attention
# Segmentation: 2s non-overlapping (best from Phase 1 Test 1)
# Features: All 41 features (best from Phase 1 Test 2)
#
# Checkpoint system: results are saved to Google Drive after every fold.
# If the session drops, re-run the script — it will resume automatically
# from the last completed fold.
#
# All outputs (PNGs, CSVs) are saved directly to OUTPUT_DIR on Drive.
# =============================================================================

# Required installs (run once in Colab if not already installed):
# !pip install librosa parselmouth soundfile scikit-learn pandas matplotlib seaborn
!pip install tensorflow

# Mount Google Drive first — run this cell before the script:
# from google.colab import drive
# drive.mount('/content/drive')

# Keepalive — run this in a separate cell to prevent idle disconnection:
# %%javascript
# window.keepAliveInterval = setInterval(function(){
#     console.log("Keeping alive...");
#     document.querySelector("colab-toolbar-button#connect").click()
# }, 60000)
#
# To stop it after the script finishes, run:
# %%javascript
# clearInterval(window.keepAliveInterval)

import os
import time
import pickle
import numpy as np
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import parselmouth
import soundfile as sf
from parselmouth.praat import call

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score,
    recall_score, confusion_matrix, classification_report
)

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Dense, Dropout, BatchNormalization,
    Conv1D, MaxPooling1D, LSTM, GlobalAveragePooling1D,
    Multiply, Activation, Flatten
)
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical

# =============================================================================
# CONFIGURATION
# =============================================================================

DATASET_PATH        = "/content/drive/My Drive/Neuroscience/Research/ravdess_male"
SAMPLE_RATE         = 16000
WINDOW_DURATION_SEC = 2
CONDITION_LABEL     = "2s-Windows-Non-Overlapping | All-41-Features"
RANDOM_STATE        = 42

VALIDATION_SPLIT    = 0.15
NEURAL_EPOCHS       = 50
NEURAL_BATCH_SIZE   = 32
EARLY_STOP_PATIENCE = 7

# All PNGs and CSVs are written here — directly on Drive, safe across sessions
OUTPUT_DIR      = "/content/drive/My Drive/DAVIS/DAVIS2_results/P2b/"
CHECKPOINT_PATH = "/content/drive/My Drive/DAVIS/DAVIS2_results/P2b_checkpoint.pkl"

emotion_map = {
    "01": "Neutral", "02": "Calm",    "03": "Happy",    "04": "Sad",
    "05": "Angry",   "06": "Fearful", "07": "Disgust",  "08": "Surprised",
}
EMOTION_LABELS = list(emotion_map.values())
N_CLASSES      = len(EMOTION_LABELS)

# =============================================================================
# OUTPUT HELPER
# =============================================================================

def out(filename):
    """Return full Drive path for an output file and ensure directory exists."""
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    return os.path.join(OUTPUT_DIR, filename)

# =============================================================================
# CHECKPOINT HELPERS
# =============================================================================

def save_checkpoint(checkpoint):
    os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)
    with open(CHECKPOINT_PATH, 'wb') as f:
        pickle.dump(checkpoint, f)
    print(f"  Checkpoint saved → {CHECKPOINT_PATH}")


def load_checkpoint():
    if os.path.exists(CHECKPOINT_PATH):
        with open(CHECKPOINT_PATH, 'rb') as f:
            checkpoint = pickle.load(f)
        completed = checkpoint.get('completed_folds', [])
        print(f"Checkpoint found — {len(completed)} fold(s) already completed: {completed}")
        print("Resuming from where session left off.\n")
        return checkpoint, len(completed)
    else:
        print("No checkpoint found — starting fresh.\n")
        return None, 0


def init_results(model_names):
    return {
        name: {
            'y_true':       [],
            'y_pred':       [],
            'fold_metrics': [],
            'train_times':  [],
            'histories':    [],
        }
        for name in model_names
    }

# =============================================================================
# FEATURE EXTRACTION  (identical pipeline to Phase 1 and 2a)
# =============================================================================

def extract_parselmouth_features(y, sr):
    temp_filename = "temp_praat.wav"
    try:
        sf.write(temp_filename, y, sr)
        snd           = parselmouth.Sound(temp_filename)
        point_process = call(snd, "To PointProcess (periodic, cc)", 75, 500)
        jitter        = call(point_process, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
        shimmer       = call([snd, point_process], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
        harmonicity   = call(snd, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
        hnr           = call(harmonicity, "Get mean", 0, 0)
        jitter  = 0.0 if (jitter  is None or not np.isfinite(jitter))  else float(jitter)
        shimmer = 0.0 if (shimmer is None or not np.isfinite(shimmer)) else float(shimmer)
        hnr     = 0.0 if (hnr     is None or not np.isfinite(hnr))     else float(hnr)
    except Exception:
        jitter, shimmer, hnr = 0.0, 0.0, 0.0
    finally:
        if os.path.exists(temp_filename):
            os.remove(temp_filename)
    return [jitter, shimmer, hnr]


def extract_features(y_clip, sr):
    MIN_SAMPLES = int(0.1 * sr)
    if len(y_clip) < MIN_SAMPLES or np.max(np.abs(y_clip)) < 1e-6:
        return None

    mfccs      = librosa.feature.mfcc(y=y_clip, sr=sr, n_mfcc=13)
    mfccs_mean = np.mean(mfccs, axis=1)
    mfccs_std  = np.std(mfccs, axis=1)

    pitches, magnitudes = librosa.piptrack(y=y_clip, sr=sr)
    med = np.median(magnitudes)
    valid_pitches = pitches[magnitudes > med] if med > 0 else np.array([])
    valid_pitches = valid_pitches[valid_pitches > 0]
    pitch_mean  = float(np.mean(valid_pitches))  if len(valid_pitches) > 0 else 0.0
    pitch_std   = float(np.std(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_min   = float(np.min(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_max   = float(np.max(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_range = pitch_max - pitch_min

    energy       = librosa.feature.rms(y=y_clip)[0]
    energy_mean  = np.mean(energy)
    energy_std   = np.std(energy)
    energy_min   = np.min(energy)
    energy_max   = np.max(energy)
    energy_range = energy_max - energy_min

    zcr      = librosa.feature.zero_crossing_rate(y_clip)[0]
    zcr_mean = np.mean(zcr)
    zcr_std  = np.std(zcr)

    jitter, shimmer, hnr = extract_parselmouth_features(y_clip, sr)

    features = np.hstack([
        mfccs_mean, mfccs_std,
        [pitch_mean, pitch_std, pitch_min, pitch_max, pitch_range],
        [energy_mean, energy_std, energy_min, energy_max, energy_range],
        [zcr_mean, zcr_std],
        [jitter, shimmer, hnr]
    ])
    return np.where(np.isfinite(features), features, 0.0)


def load_dataset(dataset_path, sr):
    X, y, actors  = [], [], []
    skipped_clips = 0
    files = sorted([f for f in os.listdir(dataset_path) if f.endswith(".wav")])
    print(f"Loading {len(files)} files | {CONDITION_LABEL}")

    for i, file in enumerate(files):
        if i % 20 == 0:
            print(f"  File {i+1}/{len(files)}...")
        parts      = file.split('-')
        emotion_id = parts[2]
        actor_id   = parts[-1].split('.')[0]
        emotion    = emotion_map.get(emotion_id, "Unknown")
        if emotion == "Unknown":
            continue

        y_full, _ = librosa.load(os.path.join(dataset_path, file), sr=sr)
        spw = int(WINDOW_DURATION_SEC * sr)
        for w in range(len(y_full) // spw):
            clip     = y_full[w*spw:(w+1)*spw]
            features = extract_features(clip, sr)
            if features is None:
                skipped_clips += 1
                continue
            X.append(features)
            y.append(emotion)
            actors.append(actor_id)

    print(f"Loaded: {len(X)} samples | Skipped: {skipped_clips}\n")
    return np.array(X), np.array(y), np.array(actors)

# =============================================================================
# MODEL DEFINITIONS
# =============================================================================

NEEDS_RESHAPE = {
    "MLP_1Layer":         False,
    "MLP_3Layer":         False,
    "CNN_1D":             True,
    "CNN_LSTM_Attention": True,
}


def build_mlp_1layer(input_dim, n_classes):
    inputs = Input(shape=(input_dim,))
    x = Dense(128, activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    outputs = Dense(n_classes, activation='softmax')(x)
    return Model(inputs, outputs, name="MLP_1Layer")


def build_mlp_3layer(input_dim, n_classes):
    inputs = Input(shape=(input_dim,))
    x = Dense(256, activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Dense(128, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Dense(64, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    outputs = Dense(n_classes, activation='softmax')(x)
    return Model(inputs, outputs, name="MLP_3Layer")


def build_cnn_1d(input_dim, n_classes):
    inputs = Input(shape=(input_dim, 1))
    x = Conv1D(64, kernel_size=3, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    x = Conv1D(128, kernel_size=3, activation='relu', padding='same')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(n_classes, activation='softmax')(x)
    return Model(inputs, outputs, name="CNN_1D")


def build_cnn_lstm(input_dim, n_classes):
    inputs = Input(shape=(input_dim, 1))
    x = Conv1D(64,  kernel_size=3, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    x = Conv1D(128, kernel_size=3, activation='relu', padding='same')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = LSTM(128, return_sequences=True)(x)
    attention = Dense(1, activation='tanh')(x)
    attention = Activation('softmax')(attention)
    x = Multiply()([x, attention])
    x = GlobalAveragePooling1D()(x)
    outputs = Dense(n_classes, activation='softmax')(x)
    return Model(inputs, outputs, name="CNN_LSTM_Attention")


NEURAL_BUILDERS = {
    "MLP_1Layer":         build_mlp_1layer,
    "MLP_3Layer":         build_mlp_3layer,
    "CNN_1D":             build_cnn_1d,
    "CNN_LSTM_Attention": build_cnn_lstm,
}

# =============================================================================
# LOOCV — ALL NEURAL MODELS INSIDE EACH FOLD, WITH CHECKPOINT/RESUME
# =============================================================================

def run_loocv(X, y, actors):
    unique_actors = np.unique(actors)
    n_folds       = len(unique_actors)
    input_dim     = X.shape[1]

    le = LabelEncoder()
    le.fit(EMOTION_LABELS)

    checkpoint, start_fold = load_checkpoint()

    if checkpoint is not None:
        results         = checkpoint['results']
        completed_folds = checkpoint['completed_folds']
    else:
        results         = init_results(NEURAL_BUILDERS.keys())
        completed_folds = []

    print("="*70)
    print(f"PHASE 2b — Neural LOOCV | {n_folds} folds total")
    print(f"{CONDITION_LABEL}")
    if start_fold > 0:
        print(f"Resuming from fold {start_fold + 1} of {n_folds}")
    print("="*70)

    for fold_num, test_actor in enumerate(unique_actors, start=1):
        if test_actor in completed_folds:
            print(f"\n--- Fold {fold_num}/{n_folds} | Actor {test_actor} — already done, skipping ---")
            continue

        print(f"\n--- Fold {fold_num}/{n_folds} | Test actor: {test_actor} ---")

        train_idx = np.where(actors != test_actor)[0]
        test_idx  = np.where(actors == test_actor)[0]

        X_train_full, X_test = X[train_idx], X[test_idx]
        y_train_full, y_test = y[train_idx], y[test_idx]

        scaler     = StandardScaler()
        X_train_sc = scaler.fit_transform(X_train_full)
        X_test_sc  = scaler.transform(X_test)

        X_tr, X_val, y_tr, y_val = train_test_split(
            X_train_sc, y_train_full,
            test_size=VALIDATION_SPLIT,
            stratify=y_train_full,
            random_state=RANDOM_STATE
        )

        y_tr_enc  = to_categorical(le.transform(y_tr),  N_CLASSES)
        y_val_enc = to_categorical(le.transform(y_val), N_CLASSES)

        X_tr_3d   = X_tr[:, :, np.newaxis]
        X_val_3d  = X_val[:, :, np.newaxis]
        X_test_3d = X_test_sc[:, :, np.newaxis]

        for model_name, builder in NEURAL_BUILDERS.items():
            print(f"  Training {model_name}...", end=' ', flush=True)

            tf.keras.backend.clear_session()
            model = builder(input_dim, N_CLASSES)
            model.compile(
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                loss='categorical_crossentropy',
                metrics=['accuracy']
            )

            if NEEDS_RESHAPE[model_name]:
                X_tr_in, X_val_in, X_test_in = X_tr_3d, X_val_3d, X_test_3d
            else:
                X_tr_in, X_val_in, X_test_in = X_tr, X_val, X_test_sc

            early_stop = EarlyStopping(
                monitor='val_loss',
                patience=EARLY_STOP_PATIENCE,
                restore_best_weights=True,
                verbose=0
            )

            t0 = time.time()
            history = model.fit(
                X_tr_in, y_tr_enc,
                validation_data=(X_val_in, y_val_enc),
                epochs=NEURAL_EPOCHS,
                batch_size=NEURAL_BATCH_SIZE,
                callbacks=[early_stop],
                verbose=0
            )
            elapsed = time.time() - t0

            y_pred_prob = model.predict(X_test_in, verbose=0)
            y_pred      = le.inverse_transform(np.argmax(y_pred_prob, axis=1))

            acc  = accuracy_score(y_test, y_pred)
            f1   = f1_score(y_test, y_pred, average='macro', zero_division=0)
            prec = precision_score(y_test, y_pred, average='macro', zero_division=0)
            rec  = recall_score(y_test, y_pred, average='macro', zero_division=0)
            n_ep = len(history.history['loss'])

            results[model_name]['y_true'].extend(y_test)
            results[model_name]['y_pred'].extend(y_pred)
            results[model_name]['train_times'].append(elapsed)
            results[model_name]['histories'].append(history.history)
            results[model_name]['fold_metrics'].append({
                'actor':      test_actor,
                'n_train':    len(y_tr),
                'n_val':      len(y_val),
                'n_test':     len(y_test),
                'accuracy':   acc,
                'macro_f1':   f1,
                'precision':  prec,
                'recall':     rec,
                'epochs_run': n_ep,
                'n_params':   model.count_params(),
            })

            print(f"Acc: {acc:.3f} | F1: {f1:.3f} | "
                  f"Epochs: {n_ep}/{NEURAL_EPOCHS} | "
                  f"Params: {model.count_params():,} | "
                  f"Time: {elapsed:.1f}s")

        completed_folds.append(test_actor)
        save_checkpoint({'results': results, 'completed_folds': completed_folds})

    return results, le

# =============================================================================
# AGGREGATION
# =============================================================================

def aggregate_results(results):
    rows = []
    for model_name, data in results.items():
        accs   = [m['accuracy']   for m in data['fold_metrics']]
        f1s    = [m['macro_f1']   for m in data['fold_metrics']]
        precs  = [m['precision']  for m in data['fold_metrics']]
        recs   = [m['recall']     for m in data['fold_metrics']]
        epochs = [m['epochs_run'] for m in data['fold_metrics']]
        n_p    = data['fold_metrics'][0].get('n_params', 'N/A')
        rows.append({
            'model':           model_name,
            'mean_accuracy':   np.mean(accs),
            'std_accuracy':    np.std(accs),
            'mean_macro_f1':   np.mean(f1s),
            'std_macro_f1':    np.std(f1s),
            'mean_precision':  np.mean(precs),
            'std_precision':   np.std(precs),
            'mean_recall':     np.mean(recs),
            'std_recall':      np.std(recs),
            'mean_epochs':     np.mean(epochs),
            'mean_train_time': np.mean(data['train_times']),
            'n_params':        n_p,
        })
    return pd.DataFrame(rows).sort_values('mean_accuracy', ascending=False).reset_index(drop=True)

# =============================================================================
# PLOTTING
# =============================================================================

def _average_history(histories, key):
    max_len = max(len(h[key]) for h in histories)
    padded  = []
    for h in histories:
        vals = list(h[key])
        pad  = [vals[-1]] * (max_len - len(vals))
        padded.append(vals + pad)
    arr = np.array(padded)
    return np.mean(arr, axis=0), np.std(arr, axis=0)


def plot_training_curves(results):
    for model_name, data in results.items():
        histories = data['histories']
        if not histories:
            continue

        mean_loss,     std_loss     = _average_history(histories, 'loss')
        mean_val_loss, std_val_loss = _average_history(histories, 'val_loss')
        mean_acc,      std_acc      = _average_history(histories, 'accuracy')
        mean_val_acc,  std_val_acc  = _average_history(histories, 'val_accuracy')
        epochs  = np.arange(1, len(mean_loss) + 1)
        n_folds = len(histories)

        fig, axes = plt.subplots(1, 2, figsize=(14, 5))

        axes[0].plot(epochs, mean_loss,     color='steelblue',  label='Train Loss',      linewidth=2)
        axes[0].fill_between(epochs, mean_loss - std_loss, mean_loss + std_loss,
                             alpha=0.2, color='steelblue')
        axes[0].plot(epochs, mean_val_loss, color='darkorange', label='Validation Loss', linewidth=2)
        axes[0].fill_between(epochs, mean_val_loss - std_val_loss, mean_val_loss + std_val_loss,
                             alpha=0.2, color='darkorange')
        axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Loss")
        axes[0].set_title(f"Loss Curves\nPhase 2b | {model_name}", fontsize=11, fontweight='bold')
        axes[0].legend(); axes[0].grid(alpha=0.3)

        axes[1].plot(epochs, mean_acc,     color='steelblue',  label='Train Accuracy',      linewidth=2)
        axes[1].fill_between(epochs, mean_acc - std_acc, mean_acc + std_acc,
                             alpha=0.2, color='steelblue')
        axes[1].plot(epochs, mean_val_acc, color='darkorange', label='Validation Accuracy', linewidth=2)
        axes[1].fill_between(epochs, mean_val_acc - std_val_acc, mean_val_acc + std_val_acc,
                             alpha=0.2, color='darkorange')
        axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Accuracy")
        axes[1].set_ylim(0, 1.0)
        axes[1].set_title(f"Accuracy Curves\nPhase 2b | {model_name}", fontsize=11, fontweight='bold')
        axes[1].legend(); axes[1].grid(alpha=0.3)

        fig.suptitle(
            f"{model_name} — Averaged across {n_folds} LOOCV folds (shaded = ±1 std)\n"
            f"{CONDITION_LABEL}", fontsize=10, y=1.01
        )
        plt.tight_layout()
        fname = out(f"P2b_training_curves_{model_name}.png")
        plt.savefig(fname, dpi=150, bbox_inches='tight')
        plt.show()
        print(f"Saved: {fname}")


def plot_model_comparison(df_summary):
    models   = df_summary['model'].tolist()
    accs     = df_summary['mean_accuracy'].tolist()
    acc_errs = df_summary['std_accuracy'].tolist()
    f1s      = df_summary['mean_macro_f1'].tolist()
    f1_errs  = df_summary['std_macro_f1'].tolist()
    x        = np.arange(len(models))
    width    = 0.35

    fig, ax = plt.subplots(figsize=(11, 6))
    b1 = ax.bar(x - width/2, accs, width, yerr=acc_errs, capsize=5,
                color='steelblue', alpha=0.85, label='Mean Accuracy')
    b2 = ax.bar(x + width/2, f1s,  width, yerr=f1_errs,  capsize=5,
                color='darkorange', alpha=0.85, label='Mean Macro-F1')

    for bar in list(b1) + list(b2):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9)

    ax.axhline(0.125, color='grey', linestyle=':', linewidth=1, label='Chance level (12.5%)')
    ax.set_xticks(x)
    ax.set_xticklabels(models, fontsize=11)
    ax.set_ylabel('Score')
    ax.set_ylim(0, min(1.0, max(accs + f1s) + 0.18))
    ax.set_title(
        "Phase 2b — Neural Model Comparison: Mean Accuracy & Macro-F1\n"
        f"{CONDITION_LABEL}", fontsize=11, fontweight='bold'
    )
    ax.legend(fontsize=10)
    plt.tight_layout()
    fname = out("P2b_neural_model_comparison.png")
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {fname}")


def plot_per_fold(results):
    n_models = len(results)
    fig, axes = plt.subplots(1, n_models, figsize=(5 * n_models, 5), sharey=True)
    if n_models == 1:
        axes = [axes]

    for ax, (model_name, data) in zip(axes, results.items()):
        actors = [m['actor']    for m in data['fold_metrics']]
        accs   = [m['accuracy'] for m in data['fold_metrics']]
        f1s    = [m['macro_f1'] for m in data['fold_metrics']]
        x      = np.arange(len(actors))

        ax.plot(x, accs, marker='o', color='steelblue',  label='Accuracy', linewidth=1.5)
        ax.plot(x, f1s,  marker='s', color='darkorange', label='Macro-F1', linewidth=1.5)
        ax.axhline(np.mean(accs), color='steelblue',  linestyle='--', linewidth=1,
                   label=f'Mean Acc={np.mean(accs):.3f}')
        ax.axhline(np.mean(f1s),  color='darkorange', linestyle='--', linewidth=1,
                   label=f'Mean F1={np.mean(f1s):.3f}')
        ax.axhline(0.125, color='grey', linestyle=':', linewidth=1)
        ax.set_xticks(x)
        ax.set_xticklabels([f'A{a}' for a in actors], rotation=45, fontsize=8)
        ax.set_title(model_name, fontsize=10, fontweight='bold')
        ax.set_ylim(0, 1.0)
        ax.legend(fontsize=7, loc='upper right')
        ax.grid(alpha=0.3)

    axes[0].set_ylabel('Score')
    fig.suptitle(
        "Phase 2b — Per-Fold Performance by Actor\n"
        f"{CONDITION_LABEL}", fontsize=11, fontweight='bold'
    )
    plt.tight_layout()
    fname = out("P2b_per_fold_all_models.png")
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {fname}")


def plot_confusion_matrix(y_true, y_pred, model_name):
    cm      = confusion_matrix(y_true, y_pred, labels=EMOTION_LABELS)
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
    cm_norm = np.nan_to_num(cm_norm)

    fig, axes = plt.subplots(1, 2, figsize=(18, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=EMOTION_LABELS, yticklabels=EMOTION_LABELS, ax=axes[0])
    axes[0].set_title(f"Confusion Matrix — Raw Counts\nPhase 2b | {model_name}",
                      fontsize=11, fontweight='bold')
    axes[0].set_xlabel("Predicted"); axes[0].set_ylabel("True")
    axes[0].tick_params(axis='x', rotation=45)

    sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=EMOTION_LABELS, yticklabels=EMOTION_LABELS,
                vmin=0, vmax=1, ax=axes[1])
    axes[1].set_title(f"Confusion Matrix — Row-Normalised\nPhase 2b | {model_name}",
                      fontsize=11, fontweight='bold')
    axes[1].set_xlabel("Predicted"); axes[1].set_ylabel("True")
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    fname = out(f"P2b_CM_{model_name}.png")
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {fname}")


def plot_summary_table(df_summary):
    col_labels = ['Model', 'Mean Acc', 'Std Acc', 'Mean F1', 'Std F1',
                  'Precision', 'Recall', 'Mean Epochs', 'Train Time (s)', 'Params']
    table_data = [
        [
            row['model'],
            f"{row['mean_accuracy']:.4f}",
            f"{row['std_accuracy']:.4f}",
            f"{row['mean_macro_f1']:.4f}",
            f"{row['std_macro_f1']:.4f}",
            f"{row['mean_precision']:.4f}",
            f"{row['mean_recall']:.4f}",
            f"{row['mean_epochs']:.1f}",
            f"{row['mean_train_time']:.1f}",
            str(row['n_params']),
        ]
        for _, row in df_summary.iterrows()
    ]

    fig, ax = plt.subplots(figsize=(18, 2.5))
    ax.axis('off')
    tbl = ax.table(cellText=table_data, colLabels=col_labels,
                   cellLoc='center', loc='center')
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(8.5)
    tbl.scale(1.1, 1.8)

    for j in range(len(col_labels)):
        tbl[0, j].set_facecolor('#2c5f8a')
        tbl[0, j].set_text_props(color='white', fontweight='bold')

    best_row = df_summary['mean_accuracy'].idxmax() + 1
    for j in range(len(col_labels)):
        tbl[best_row, j].set_facecolor('#d4edff')

    ax.set_title(
        "Phase 2b — Neural Models Results Summary\n"
        f"{CONDITION_LABEL}  (blue = best accuracy)",
        fontsize=10, fontweight='bold', pad=14
    )
    plt.tight_layout()
    fname = out("P2b_results_table.png")
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {fname}")

# =============================================================================
# MAIN
# =============================================================================

if __name__ == "__main__":
    # 1. Load data
    X, y, actors = load_dataset(DATASET_PATH, SAMPLE_RATE)
    print(f"Feature matrix : {X.shape}")
    print(f"NaN: {int(np.isnan(X).sum())} | Inf: {int(np.isinf(X).sum())}")
    print(f"Class distribution:\n{pd.Series(y).value_counts().sort_index().to_string()}\n")

    # 2. LOOCV with checkpoint/resume
    results, le = run_loocv(X, y, actors)

    # 3. Aggregate
    df_summary = aggregate_results(results)

    # 4. Print summary
    print("\n" + "="*70)
    print("PHASE 2b RESULTS SUMMARY — Neural Models")
    print("="*70)
    print(df_summary.to_string(index=False))

    # 5. Per-class reports
    print("\n" + "="*70)
    print("PER-CLASS CLASSIFICATION REPORTS")
    print("="*70)
    for model_name, data in results.items():
        print(f"\n--- {model_name} ---")
        print(classification_report(
            data['y_true'], data['y_pred'],
            labels=EMOTION_LABELS, target_names=EMOTION_LABELS, zero_division=0
        ))

    # 6. Plots — all saved directly to Drive via out()
    plot_model_comparison(df_summary)
    plot_per_fold(results)
    plot_training_curves(results)
    for model_name, data in results.items():
        plot_confusion_matrix(data['y_true'], data['y_pred'], model_name)
    plot_summary_table(df_summary)

    # 7. Save CSVs directly to Drive
    fpath = out("P2b_summary.csv")
    df_summary.to_csv(fpath, index=False)
    print(f"Saved: {fpath}")

    all_fold_rows = []
    for model_name, data in results.items():
        for m in data['fold_metrics']:
            m['model'] = model_name
            all_fold_rows.append(m)
    fpath = out("P2b_fold_metrics.csv")
    pd.DataFrame(all_fold_rows).to_csv(fpath, index=False)
    print(f"Saved: {fpath}")

    # 8. Clean up checkpoint
    if os.path.exists(CHECKPOINT_PATH):
        os.remove(CHECKPOINT_PATH)
        print("Checkpoint removed — run complete.")

    print("\nDone. Phase 2b complete.")
```
