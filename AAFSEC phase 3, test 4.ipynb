```python
# =============================================================================
# DAVIS - Phase 3, Test 4: Emotional Intensity Analysis
# Goal: Determine whether emotional intensity (low vs high) affects
#       classification performance across selected models
# Models: Random Forest, Logistic Regression, MLP_3Layer, CNN_LSTM_Attention
# Segmentation: 2s non-overlapping (locked from Phase 1)
# Features: All 41 features (locked from Phase 1)
#
# RAVDESS intensity coding (filename position 4, 1-indexed):
#   "01" = normal/low intensity
#   "02" = strong/high intensity
#   Note: Neutral only has normal intensity — it is excluded from this
#         analysis to keep low/high comparisons clean and balanced.
#
# All outputs saved directly to OUTPUT_DIR on Drive.
# =============================================================================

# Required installs (run once in Colab if not already installed):
!pip install librosa praat-parselmouth soundfile scikit-learn pandas matplotlib seaborn
!pip install tensorflow

# Mount Google Drive first:
# from google.colab import drive
# drive.mount('/content/drive')

import os
import time
import numpy as np
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import parselmouth
import soundfile as sf
from parselmouth.praat import call

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score,
    recall_score, confusion_matrix, classification_report
)

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Dense, Dropout, BatchNormalization,
    Conv1D, MaxPooling1D, LSTM, GlobalAveragePooling1D,
    Multiply, Activation, Flatten
)
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical

# =============================================================================
# CONFIGURATION
# =============================================================================

DATASET_PATH        = "/content/drive/My Drive/Neuroscience/Research/ravdess_male"
SAMPLE_RATE         = 16000
WINDOW_DURATION_SEC = 2
RANDOM_STATE        = 42
VALIDATION_SPLIT    = 0.15
NEURAL_EPOCHS       = 50
NEURAL_BATCH_SIZE   = 32
EARLY_STOP_PATIENCE = 7

OUTPUT_DIR = "/content/drive/My Drive/DAVIS/DAVIS2_results/P3T4/"

# Emotions included in intensity analysis — Neutral excluded (no high intensity)
INTENSITY_EMOTIONS = ["Calm", "Happy", "Sad", "Angry", "Fearful", "Disgust", "Surprised"]
N_CLASSES          = len(INTENSITY_EMOTIONS)

emotion_map = {
    "01": "Neutral", "02": "Calm",    "03": "Happy",    "04": "Sad",
    "05": "Angry",   "06": "Fearful", "07": "Disgust",  "08": "Surprised",
}

# RAVDESS filename structure (1-indexed, dash-separated):
# Position 4 = intensity: "01" = low, "02" = high
intensity_map = {"01": "low", "02": "high"}

# =============================================================================
# OUTPUT HELPER
# =============================================================================

def out(filename):
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    return os.path.join(OUTPUT_DIR, filename)

# =============================================================================
# FEATURE EXTRACTION  (identical pipeline throughout)
# =============================================================================

def extract_parselmouth_features(y, sr):
    temp_filename = "temp_praat.wav"
    try:
        sf.write(temp_filename, y, sr)
        snd           = parselmouth.Sound(temp_filename)
        point_process = call(snd, "To PointProcess (periodic, cc)", 75, 500)
        jitter        = call(point_process, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
        shimmer       = call([snd, point_process], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
        harmonicity   = call(snd, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
        hnr           = call(harmonicity, "Get mean", 0, 0)
        jitter  = 0.0 if (jitter  is None or not np.isfinite(jitter))  else float(jitter)
        shimmer = 0.0 if (shimmer is None or not np.isfinite(shimmer)) else float(shimmer)
        hnr     = 0.0 if (hnr     is None or not np.isfinite(hnr))     else float(hnr)
    except Exception:
        jitter, shimmer, hnr = 0.0, 0.0, 0.0
    finally:
        if os.path.exists(temp_filename):
            os.remove(temp_filename)
    return [jitter, shimmer, hnr]


def extract_features(y_clip, sr):
    MIN_SAMPLES = int(0.1 * sr)
    if len(y_clip) < MIN_SAMPLES or np.max(np.abs(y_clip)) < 1e-6:
        return None

    mfccs      = librosa.feature.mfcc(y=y_clip, sr=sr, n_mfcc=13)
    mfccs_mean = np.mean(mfccs, axis=1)
    mfccs_std  = np.std(mfccs, axis=1)

    pitches, magnitudes = librosa.piptrack(y=y_clip, sr=sr)
    med = np.median(magnitudes)
    valid_pitches = pitches[magnitudes > med] if med > 0 else np.array([])
    valid_pitches = valid_pitches[valid_pitches > 0]
    pitch_mean  = float(np.mean(valid_pitches))  if len(valid_pitches) > 0 else 0.0
    pitch_std   = float(np.std(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_min   = float(np.min(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_max   = float(np.max(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_range = pitch_max - pitch_min

    energy       = librosa.feature.rms(y=y_clip)[0]
    energy_mean  = np.mean(energy)
    energy_std   = np.std(energy)
    energy_min   = np.min(energy)
    energy_max   = np.max(energy)
    energy_range = energy_max - energy_min

    zcr      = librosa.feature.zero_crossing_rate(y_clip)[0]
    zcr_mean = np.mean(zcr)
    zcr_std  = np.std(zcr)

    jitter, shimmer, hnr = extract_parselmouth_features(y_clip, sr)

    features = np.hstack([
        mfccs_mean, mfccs_std,
        [pitch_mean, pitch_std, pitch_min, pitch_max, pitch_range],
        [energy_mean, energy_std, energy_min, energy_max, energy_range],
        [zcr_mean, zcr_std],
        [jitter, shimmer, hnr]
    ])
    return np.where(np.isfinite(features), features, 0.0)

# =============================================================================
# DATA LOADING — with intensity label extracted from filename
# =============================================================================

def load_dataset(dataset_path, sr):
    """
    Load all non-Neutral RAVDESS files, segmented into 2s windows.
    Returns X, y (emotion), actors, intensities ("low" or "high").
    Neutral is excluded since it has no high-intensity variant.
    """
    X, y, actors, intensities = [], [], [], []
    skipped = 0

    files = sorted([f for f in os.listdir(dataset_path) if f.endswith(".wav")])
    print(f"Loading {len(files)} files...")

    for i, file in enumerate(files):
        if i % 20 == 0:
            print(f"  File {i+1}/{len(files)}...")

        parts        = file.split('-')
        emotion_id   = parts[2]
        intensity_id = parts[3]   # position 4 in filename (1-indexed), 0-indexed = 3
        actor_id     = parts[-1].split('.')[0]

        emotion   = emotion_map.get(emotion_id, "Unknown")
        intensity = intensity_map.get(intensity_id, "Unknown")

        # Skip Neutral and any unknown labels
        if emotion not in INTENSITY_EMOTIONS or intensity == "Unknown":
            continue

        y_full, _ = librosa.load(os.path.join(dataset_path, file), sr=sr)
        spw = int(WINDOW_DURATION_SEC * sr)

        for w in range(len(y_full) // spw):
            clip     = y_full[w*spw:(w+1)*spw]
            features = extract_features(clip, sr)
            if features is None:
                skipped += 1
                continue
            X.append(features)
            y.append(emotion)
            actors.append(actor_id)
            intensities.append(intensity)

    print(f"Loaded: {len(X)} samples | Skipped: {skipped} invalid clips")
    print(f"Low intensity:  {intensities.count('low')}")
    print(f"High intensity: {intensities.count('high')}\n")

    return (
        np.array(X),
        np.array(y),
        np.array(actors),
        np.array(intensities)
    )

# =============================================================================
# MODEL BUILDERS
# =============================================================================

def build_mlp_3layer(input_dim, n_classes):
    inputs = Input(shape=(input_dim,))
    x = Dense(256, activation='relu')(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Dense(128, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    x = Dense(64, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    outputs = Dense(n_classes, activation='softmax')(x)
    return Model(inputs, outputs, name="MLP_3Layer")


def build_cnn_lstm(input_dim, n_classes):
    inputs = Input(shape=(input_dim, 1))
    x = Conv1D(64,  kernel_size=3, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(pool_size=2)(x)
    x = Conv1D(128, kernel_size=3, activation='relu', padding='same')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = LSTM(128, return_sequences=True)(x)
    attention = Dense(1, activation='tanh')(x)
    attention = Activation('softmax')(attention)
    x = Multiply()([x, attention])
    x = GlobalAveragePooling1D()(x)
    outputs = Dense(n_classes, activation='softmax')(x)
    return Model(inputs, outputs, name="CNN_LSTM_Attention")

# =============================================================================
# LOOCV FOR A SINGLE MODEL + INTENSITY CONDITION
# =============================================================================

def run_loocv_classical(X, y, actors, model_fn, model_name, condition_label):
    unique_actors = np.unique(actors)
    all_y_true, all_y_pred = [], []
    fold_metrics = []

    for test_actor in unique_actors:
        train_idx = np.where(actors != test_actor)[0]
        test_idx  = np.where(actors == test_actor)[0]

        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        scaler     = StandardScaler()
        X_train_sc = scaler.fit_transform(X_train)
        X_test_sc  = scaler.transform(X_test)

        model = model_fn()
        model.fit(X_train_sc, y_train)
        y_pred = model.predict(X_test_sc)

        fold_metrics.append({
            'actor':    test_actor,
            'accuracy': accuracy_score(y_test, y_pred),
            'macro_f1': f1_score(y_test, y_pred, average='macro', zero_division=0),
        })
        all_y_true.extend(y_test)
        all_y_pred.extend(y_pred)

    accs = [m['accuracy'] for m in fold_metrics]
    f1s  = [m['macro_f1'] for m in fold_metrics]

    summary = {
        'model':           model_name,
        'condition':       condition_label,
        'mean_accuracy':   np.mean(accs),
        'std_accuracy':    np.std(accs),
        'mean_macro_f1':   np.mean(f1s),
        'std_macro_f1':    np.std(f1s),
        'mean_precision':  np.mean([
            precision_score(all_y_true, all_y_pred, average='macro', zero_division=0)
        ]),
        'mean_recall': np.mean([
            recall_score(all_y_true, all_y_pred, average='macro', zero_division=0)
        ]),
    }
    return summary, np.array(all_y_true), np.array(all_y_pred)


def run_loocv_neural(X, y, actors, builder, needs_reshape,
                     model_name, condition_label, le):
    unique_actors = np.unique(actors)
    input_dim     = X.shape[1]
    all_y_true, all_y_pred = [], []
    fold_metrics = []

    for test_actor in unique_actors:
        train_idx = np.where(actors != test_actor)[0]
        test_idx  = np.where(actors == test_actor)[0]

        X_train_full, X_test = X[train_idx], X[test_idx]
        y_train_full, y_test = y[train_idx], y[test_idx]

        scaler     = StandardScaler()
        X_train_sc = scaler.fit_transform(X_train_full)
        X_test_sc  = scaler.transform(X_test)

        X_tr, X_val, y_tr, y_val = train_test_split(
            X_train_sc, y_train_full,
            test_size=VALIDATION_SPLIT,
            stratify=y_train_full,
            random_state=RANDOM_STATE
        )

        y_tr_enc  = to_categorical(le.transform(y_tr),  N_CLASSES)
        y_val_enc = to_categorical(le.transform(y_val), N_CLASSES)

        if needs_reshape:
            X_tr_in   = X_tr[:, :, np.newaxis]
            X_val_in  = X_val[:, :, np.newaxis]
            X_test_in = X_test_sc[:, :, np.newaxis]
        else:
            X_tr_in, X_val_in, X_test_in = X_tr, X_val, X_test_sc

        tf.keras.backend.clear_session()
        model = builder(input_dim, N_CLASSES)
        model.compile(
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        early_stop = EarlyStopping(
            monitor='val_loss', patience=EARLY_STOP_PATIENCE,
            restore_best_weights=True, verbose=0
        )
        model.fit(
            X_tr_in, y_tr_enc,
            validation_data=(X_val_in, y_val_enc),
            epochs=NEURAL_EPOCHS,
            batch_size=NEURAL_BATCH_SIZE,
            callbacks=[early_stop],
            verbose=0
        )

        y_pred = le.inverse_transform(
            np.argmax(model.predict(X_test_in, verbose=0), axis=1)
        )

        fold_metrics.append({
            'actor':    test_actor,
            'accuracy': accuracy_score(y_test, y_pred),
            'macro_f1': f1_score(y_test, y_pred, average='macro', zero_division=0),
        })
        all_y_true.extend(y_test)
        all_y_pred.extend(y_pred)

    accs = [m['accuracy'] for m in fold_metrics]
    f1s  = [m['macro_f1'] for m in fold_metrics]

    summary = {
        'model':          model_name,
        'condition':      condition_label,
        'mean_accuracy':  np.mean(accs),
        'std_accuracy':   np.std(accs),
        'mean_macro_f1':  np.mean(f1s),
        'std_macro_f1':   np.std(f1s),
        'mean_precision': precision_score(
            all_y_true, all_y_pred, average='macro', zero_division=0),
        'mean_recall':    recall_score(
            all_y_true, all_y_pred, average='macro', zero_division=0),
    }
    return summary, np.array(all_y_true), np.array(all_y_pred)

# =============================================================================
# PLOTTING
# =============================================================================

def plot_confusion_matrix(y_true, y_pred, model_name, condition_label):
    cm      = confusion_matrix(y_true, y_pred, labels=INTENSITY_EMOTIONS)
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
    cm_norm = np.nan_to_num(cm_norm)

    fig, axes = plt.subplots(1, 2, figsize=(18, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=INTENSITY_EMOTIONS, yticklabels=INTENSITY_EMOTIONS,
                ax=axes[0])
    axes[0].set_title(
        f"Confusion Matrix — Raw Counts\n"
        f"Phase 3 | Test 4 | {model_name} | {condition_label}",
        fontsize=11, fontweight='bold')
    axes[0].set_xlabel("Predicted"); axes[0].set_ylabel("True")
    axes[0].tick_params(axis='x', rotation=45)

    sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=INTENSITY_EMOTIONS, yticklabels=INTENSITY_EMOTIONS,
                vmin=0, vmax=1, ax=axes[1])
    axes[1].set_title(
        f"Confusion Matrix — Row-Normalised\n"
        f"Phase 3 | Test 4 | {model_name} | {condition_label}",
        fontsize=11, fontweight='bold')
    axes[1].set_xlabel("Predicted"); axes[1].set_ylabel("True")
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    safe_model = model_name.replace(' ', '_')
    safe_cond  = condition_label.replace(' ', '_')
    fname = out(f"P3_T4_CM_{safe_model}_{safe_cond}.png")
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {fname}")


def plot_intensity_comparison(all_summaries):
    """
    Side-by-side grouped bar chart: for each model, low vs high intensity,
    showing accuracy and macro-F1.
    """
    model_names = list(dict.fromkeys(s['model'] for s in all_summaries))
    conditions  = ['Low Intensity', 'High Intensity']
    x           = np.arange(len(model_names))
    width       = 0.18
    offsets     = [-1.5, -0.5, 0.5, 1.5]  # 4 bars per model group

    # Map: (model, condition) -> summary
    lookup = {(s['model'], s['condition']): s for s in all_summaries}

    fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=False)
    colors = {
        ('Low Intensity',  'acc'): '#4a90d9',
        ('High Intensity', 'acc'): '#1a5276',
        ('Low Intensity',  'f1'):  '#f0a500',
        ('High Intensity', 'f1'):  '#b7770d',
    }

    for ax, metric, metric_label in zip(
        axes,
        ['mean_accuracy', 'mean_macro_f1'],
        ['Mean Accuracy', 'Mean Macro-F1']
    ):
        for ci, cond in enumerate(conditions):
            vals     = [lookup.get((m, cond), {}).get(metric, 0) for m in model_names]
            err_key  = metric.replace('mean_', 'std_')
            errs     = [lookup.get((m, cond), {}).get(err_key, 0) for m in model_names]
            offset   = (ci - 0.5) * width * 2
            color    = colors[(cond, 'acc' if 'accuracy' in metric else 'f1')]
            bars     = ax.bar(x + offset, vals, width * 1.8,
                              yerr=errs, capsize=4,
                              label=cond, color=color, alpha=0.85)
            for bar in bars:
                ax.text(bar.get_x() + bar.get_width()/2,
                        bar.get_height() + 0.01,
                        f'{bar.get_height():.3f}',
                        ha='center', va='bottom', fontsize=7.5)

        ax.axhline(0.125, color='grey', linestyle=':', linewidth=1,
                   label='Chance (12.5%)')
        ax.set_xticks(x)
        ax.set_xticklabels(model_names, fontsize=9, rotation=15, ha='right')
        ax.set_ylabel(metric_label)
        ax.set_ylim(0, 0.75)
        ax.set_title(
            f"{metric_label}: Low vs High Intensity\n"
            f"Phase 3 | Test 4",
            fontsize=11, fontweight='bold'
        )
        ax.legend(fontsize=9)
        ax.grid(axis='y', alpha=0.3)

    plt.tight_layout()
    fname = out("P3_T4_intensity_comparison.png")
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {fname}")


def plot_summary_table(all_summaries):
    col_labels = ['Model', 'Condition', 'Mean Acc', 'Std Acc',
                  'Mean F1', 'Std F1', 'Precision', 'Recall']
    table_data = [
        [
            s['model'], s['condition'],
            f"{s['mean_accuracy']:.4f}",  f"{s['std_accuracy']:.4f}",
            f"{s['mean_macro_f1']:.4f}",  f"{s['std_macro_f1']:.4f}",
            f"{s['mean_precision']:.4f}", f"{s['mean_recall']:.4f}",
        ]
        for s in all_summaries
    ]

    fig, ax = plt.subplots(figsize=(16, len(all_summaries) * 0.55 + 1.5))
    ax.axis('off')
    tbl = ax.table(cellText=table_data, colLabels=col_labels,
                   cellLoc='center', loc='center')
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(9)
    tbl.scale(1.1, 1.7)

    for j in range(len(col_labels)):
        tbl[0, j].set_facecolor('#2c5f8a')
        tbl[0, j].set_text_props(color='white', fontweight='bold')

    # Alternate row shading by model for readability
    model_names = list(dict.fromkeys(s['model'] for s in all_summaries))
    for row_i, s in enumerate(all_summaries, start=1):
        shade = '#f0f4f8' if model_names.index(s['model']) % 2 == 0 else '#ffffff'
        for j in range(len(col_labels)):
            tbl[row_i, j].set_facecolor(shade)

    ax.set_title(
        "Phase 3 | Test 4 — Intensity Analysis Results Summary",
        fontsize=11, fontweight='bold', pad=14
    )
    plt.tight_layout()
    fname = out("P3_T4_results_table.png")
    plt.savefig(fname, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Saved: {fname}")

# =============================================================================
# MAIN
# =============================================================================

if __name__ == "__main__":
    # 1. Load full dataset with intensity labels
    X_all, y_all, actors_all, intensities_all = load_dataset(DATASET_PATH, SAMPLE_RATE)

    print(f"Feature matrix : {X_all.shape}")
    print(f"NaN: {int(np.isnan(X_all).sum())} | Inf: {int(np.isinf(X_all).sum())}\n")

    # 2. Split into low and high intensity subsets
    low_idx  = np.where(intensities_all == 'low')[0]
    high_idx = np.where(intensities_all == 'high')[0]

    X_low,  y_low,  actors_low  = X_all[low_idx],  y_all[low_idx],  actors_all[low_idx]
    X_high, y_high, actors_high = X_all[high_idx], y_all[high_idx], actors_all[high_idx]

    print(f"Low intensity subset  : {X_low.shape[0]} samples")
    print(f"High intensity subset : {X_high.shape[0]} samples\n")
    print(f"Low  class dist:\n{pd.Series(y_low).value_counts().sort_index().to_string()}\n")
    print(f"High class dist:\n{pd.Series(y_high).value_counts().sort_index().to_string()}\n")

    # 3. Label encoder for neural models — fitted on intensity emotions only
    le = LabelEncoder()
    le.fit(INTENSITY_EMOTIONS)

    # 4. Define models to test
    classical_model_fns = {
        "Random Forest": lambda: RandomForestClassifier(
            n_estimators=200, min_samples_leaf=2,
            random_state=RANDOM_STATE, n_jobs=-1
        ),
        "Logistic Regression": lambda: LogisticRegression(
            max_iter=1000, C=1.0, solver='lbfgs',
            multi_class='multinomial', random_state=RANDOM_STATE
        ),
    }
    neural_model_fns = {
        "MLP_3Layer":         (build_mlp_3layer,  False),
        "CNN_LSTM_Attention": (build_cnn_lstm,     True),
    }

    all_summaries  = []
    all_preds      = {}   # keyed by (model_name, condition) for CM plotting

    # 5. Run all models on both intensity conditions
    for condition_label, X_sub, y_sub, actors_sub in [
        ("Low Intensity",  X_low,  y_low,  actors_low),
        ("High Intensity", X_high, y_high, actors_high),
    ]:
        print("="*65)
        print(f"CONDITION: {condition_label}")
        print("="*65)

        # Classical
        for model_name, model_fn in classical_model_fns.items():
            print(f"  {model_name}...", end=' ', flush=True)
            summary, y_true, y_pred = run_loocv_classical(
                X_sub, y_sub, actors_sub, model_fn, model_name, condition_label
            )
            all_summaries.append(summary)
            all_preds[(model_name, condition_label)] = (y_true, y_pred)
            print(f"Acc: {summary['mean_accuracy']:.3f} | "
                  f"F1: {summary['mean_macro_f1']:.3f}")

        # Neural
        for model_name, (builder, needs_reshape) in neural_model_fns.items():
            print(f"  {model_name}...", end=' ', flush=True)
            summary, y_true, y_pred = run_loocv_neural(
                X_sub, y_sub, actors_sub,
                builder, needs_reshape,
                model_name, condition_label, le
            )
            all_summaries.append(summary)
            all_preds[(model_name, condition_label)] = (y_true, y_pred)
            print(f"Acc: {summary['mean_accuracy']:.3f} | "
                  f"F1: {summary['mean_macro_f1']:.3f}")

    # 6. Print full summary
    print("\n" + "="*65)
    print("PHASE 3 | TEST 4 — RESULTS SUMMARY")
    print("="*65)
    df_summary = pd.DataFrame(all_summaries)
    print(df_summary.to_string(index=False))

    # 7. Per-class reports
    print("\n" + "="*65)
    print("PER-CLASS CLASSIFICATION REPORTS")
    print("="*65)
    for (model_name, condition_label), (y_true, y_pred) in all_preds.items():
        print(f"\n--- {model_name} | {condition_label} ---")
        print(classification_report(
            y_true, y_pred,
            labels=INTENSITY_EMOTIONS, target_names=INTENSITY_EMOTIONS,
            zero_division=0
        ))

    # 8. Plots
    plot_intensity_comparison(all_summaries)
    plot_summary_table(all_summaries)

    for (model_name, condition_label), (y_true, y_pred) in all_preds.items():
        plot_confusion_matrix(y_true, y_pred, model_name, condition_label)

    # 9. Save CSVs
    fpath = out("P3_T4_summary.csv")
    df_summary.to_csv(fpath, index=False)
    print(f"Saved: {fpath}")

    fold_rows = []
    for (model_name, condition_label), (y_true, y_pred) in all_preds.items():
        fold_rows.append({
            'model':     model_name,
            'condition': condition_label,
            'n_samples': len(y_true),
        })
    fpath = out("P3_T4_condition_counts.csv")
    pd.DataFrame(fold_rows).to_csv(fpath, index=False)
    print(f"Saved: {fpath}")

    print("\nDone. Phase 3 Test 4 complete — run phase3_test5 next.")
```
