```python
# =============================================================================
# DAVIS - Phase 1, Test 2: Feature Group Importance
# Goal: Determine which feature groups contribute most to emotion classification
# Conditions: MFCC only | Prosody only | Voice Quality only | All features
# Segmentation: 2s non-overlapping windows (best from Test 1)
# Model: Logistic Regression with LOOCV
# =============================================================================

# Required installs (run once in Colab if not already installed):
# !pip install librosa parselmouth soundfile scikit-learn pandas matplotlib seaborn

import os
import numpy as np
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import parselmouth
import soundfile as sf
from parselmouth.praat import call
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score,
    recall_score, classification_report
)

# =============================================================================
# CONFIGURATION
# Segmentation locked to 2s non-overlapping (best from Phase 1 Test 1)
# =============================================================================

DATASET_PATH        = "/content/drive/My Drive/Neuroscience/Research/ravdess_male"
SAMPLE_RATE         = 16000
WINDOW_DURATION_SEC = 2     # locked — do not change
CONDITION_LABEL     = "2s-Windows-Non-Overlapping"   # for plot titles

emotion_map = {
    "01": "Neutral",
    "02": "Calm",
    "03": "Happy",
    "04": "Sad",
    "05": "Angry",
    "06": "Fearful",
    "07": "Disgust",
    "08": "Surprised",
}
EMOTION_LABELS = list(emotion_map.values())

# Feature group index ranges within the 41-feature vector:
#   [0:13]  — MFCC means        (13 features)
#   [13:26] — MFCC stds         (13 features)
#   [26:31] — Pitch stats        (5 features: mean, std, min, max, range)
#   [31:36] — Energy/RMS stats   (5 features: mean, std, min, max, range)
#   [36:38] — ZCR                (2 features: mean, std)
#   [38:41] — Voice quality      (3 features: jitter, shimmer, HNR)
#
# Group definitions used in this test:
#   MFCC     : indices 0–25  (MFCC means + stds)
#   Prosody  : indices 26–37 (pitch + energy + ZCR)
#   Voice Q  : indices 38–40 (jitter, shimmer, HNR)
#   All      : indices 0–40

FEATURE_GROUPS = {
    "MFCC Only":          list(range(0, 26)),
    "Prosody Only":       list(range(26, 38)),
    "Voice Quality Only": list(range(38, 41)),
    "All Features":       list(range(0, 41)),
}

# =============================================================================
# FEATURE EXTRACTION  (identical to Test 1 — full 41-feature vector)
# =============================================================================

def extract_parselmouth_features(y, sr):
    """
    Jitter, shimmer, HNR via Parselmouth.
    Returns [0, 0, 0] on failure or insufficient voicing.
    """
    temp_filename = "temp_praat.wav"
    try:
        sf.write(temp_filename, y, sr)
        snd           = parselmouth.Sound(temp_filename)
        point_process = call(snd, "To PointProcess (periodic, cc)", 75, 500)
        jitter        = call(point_process, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)
        shimmer       = call([snd, point_process], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)
        harmonicity   = call(snd, "To Harmonicity (cc)", 0.01, 75, 0.1, 1.0)
        hnr           = call(harmonicity, "Get mean", 0, 0)
        jitter  = 0.0 if (jitter  is None or not np.isfinite(jitter))  else float(jitter)
        shimmer = 0.0 if (shimmer is None or not np.isfinite(shimmer)) else float(shimmer)
        hnr     = 0.0 if (hnr     is None or not np.isfinite(hnr))     else float(hnr)
    except Exception:
        jitter, shimmer, hnr = 0.0, 0.0, 0.0
    finally:
        if os.path.exists(temp_filename):
            os.remove(temp_filename)
    return [jitter, shimmer, hnr]


def extract_features(y_clip, sr):
    """
    Returns full 41-feature vector, or None if clip is unusable.
    Index layout:
      0–12  : MFCC means
      13–25 : MFCC stds
      26–30 : pitch (mean, std, min, max, range)
      31–35 : energy/RMS (mean, std, min, max, range)
      36–37 : ZCR (mean, std)
      38–40 : jitter, shimmer, HNR
    """
    MIN_SAMPLES = int(0.1 * sr)
    if len(y_clip) < MIN_SAMPLES or np.max(np.abs(y_clip)) < 1e-6:
        return None

    # MFCCs
    mfccs      = librosa.feature.mfcc(y=y_clip, sr=sr, n_mfcc=13)
    mfccs_mean = np.mean(mfccs, axis=1)
    mfccs_std  = np.std(mfccs, axis=1)

    # Pitch
    pitches, magnitudes = librosa.piptrack(y=y_clip, sr=sr)
    med = np.median(magnitudes)
    valid_pitches = pitches[magnitudes > med] if med > 0 else np.array([])
    valid_pitches = valid_pitches[valid_pitches > 0]
    pitch_mean  = float(np.mean(valid_pitches))  if len(valid_pitches) > 0 else 0.0
    pitch_std   = float(np.std(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_min   = float(np.min(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_max   = float(np.max(valid_pitches))   if len(valid_pitches) > 0 else 0.0
    pitch_range = pitch_max - pitch_min

    # Energy (RMS)
    energy       = librosa.feature.rms(y=y_clip)[0]
    energy_mean  = np.mean(energy)
    energy_std   = np.std(energy)
    energy_min   = np.min(energy)
    energy_max   = np.max(energy)
    energy_range = energy_max - energy_min

    # ZCR
    zcr      = librosa.feature.zero_crossing_rate(y_clip)[0]
    zcr_mean = np.mean(zcr)
    zcr_std  = np.std(zcr)

    # Voice quality
    jitter, shimmer, hnr = extract_parselmouth_features(y_clip, sr)

    features = np.hstack([
        mfccs_mean, mfccs_std,                                          # 0–25
        [pitch_mean, pitch_std, pitch_min, pitch_max, pitch_range],     # 26–30
        [energy_mean, energy_std, energy_min, energy_max, energy_range],# 31–35
        [zcr_mean, zcr_std],                                            # 36–37
        [jitter, shimmer, hnr]                                          # 38–40
    ])

    features = np.where(np.isfinite(features), features, 0.0)
    return features

# =============================================================================
# DATA LOADING  (2s non-overlapping, locked)
# =============================================================================

def load_dataset(dataset_path, sr):
    X, y, actors  = [], [], []
    skipped_clips = 0

    files = sorted([f for f in os.listdir(dataset_path) if f.endswith(".wav")])
    print(f"Found {len(files)} .wav files | Segmentation: {CONDITION_LABEL}")

    for i, file in enumerate(files):
        if i % 20 == 0:
            print(f"  Processing file {i+1}/{len(files)}...")

        parts      = file.split('-')
        emotion_id = parts[2]
        actor_id   = parts[-1].split('.')[0]
        emotion    = emotion_map.get(emotion_id, "Unknown")
        if emotion == "Unknown":
            continue

        y_full, _ = librosa.load(os.path.join(dataset_path, file), sr=sr)
        samples_per_window = int(WINDOW_DURATION_SEC * sr)
        num_windows        = len(y_full) // samples_per_window

        for w in range(num_windows):
            start    = w * samples_per_window
            clip     = y_full[start:start + samples_per_window]
            features = extract_features(clip, sr)
            if features is None:
                skipped_clips += 1
                continue
            X.append(features)
            y.append(emotion)
            actors.append(actor_id)

    print(f"Loaded: {len(X)} samples | Skipped: {skipped_clips} invalid clips\n")
    return np.array(X), np.array(y), np.array(actors)

# =============================================================================
# LOOCV FOR A GIVEN FEATURE SUBSET
# =============================================================================

def run_loocv(X_full, y, actors, feature_indices, group_name):
    """
    Run LOOCV using only the columns in feature_indices.
    Returns per-fold metrics and aggregated predictions.
    """
    X = X_full[:, feature_indices]

    unique_actors = np.unique(actors)
    fold_metrics  = []
    all_y_true    = []
    all_y_pred    = []

    for test_actor in unique_actors:
        train_idx = np.where(actors != test_actor)[0]
        test_idx  = np.where(actors == test_actor)[0]

        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        scaler     = StandardScaler()
        X_train_sc = scaler.fit_transform(X_train)
        X_test_sc  = scaler.transform(X_test)

        model = LogisticRegression(
            max_iter=1000, C=1.0,
            solver='lbfgs', multi_class='multinomial',
            random_state=42
        )
        model.fit(X_train_sc, y_train)
        y_pred = model.predict(X_test_sc)

        fold_metrics.append({
            'actor':     test_actor,
            'n_train':   len(y_train),
            'n_test':    len(y_test),
            'accuracy':  accuracy_score(y_test, y_pred),
            'macro_f1':  f1_score(y_test, y_pred, average='macro', zero_division=0),
            'precision': precision_score(y_test, y_pred, average='macro', zero_division=0),
            'recall':    recall_score(y_test, y_pred, average='macro', zero_division=0),
        })
        all_y_true.extend(y_test)
        all_y_pred.extend(y_pred)

    accs = [m['accuracy'] for m in fold_metrics]
    f1s  = [m['macro_f1'] for m in fold_metrics]

    summary = {
        'feature_group':  group_name,
        'n_features':     len(feature_indices),
        'mean_accuracy':  np.mean(accs),
        'std_accuracy':   np.std(accs),
        'mean_macro_f1':  np.mean(f1s),
        'std_macro_f1':   np.std(f1s),
        'mean_precision': np.mean([m['precision'] for m in fold_metrics]),
        'mean_recall':    np.mean([m['recall']    for m in fold_metrics]),
    }

    print(f"  {group_name:<22} | "
          f"n={len(feature_indices):2d} | "
          f"Acc: {summary['mean_accuracy']:.4f} ± {summary['std_accuracy']:.4f} | "
          f"Macro-F1: {summary['mean_macro_f1']:.4f} ± {summary['std_macro_f1']:.4f}")

    return summary, fold_metrics, np.array(all_y_true), np.array(all_y_pred)

# =============================================================================
# PLOTTING
# =============================================================================

def plot_group_comparison(all_summaries):
    """
    Bar chart comparing mean accuracy and macro-F1 across feature groups,
    with std error bars.
    """
    groups   = [s['feature_group'] for s in all_summaries]
    accs     = [s['mean_accuracy'] for s in all_summaries]
    acc_errs = [s['std_accuracy']  for s in all_summaries]
    f1s      = [s['mean_macro_f1'] for s in all_summaries]
    f1_errs  = [s['std_macro_f1']  for s in all_summaries]

    x     = np.arange(len(groups))
    width = 0.35

    fig, ax = plt.subplots(figsize=(11, 6))
    bars1 = ax.bar(x - width/2, accs, width, yerr=acc_errs, capsize=5,
                   label='Mean Accuracy', color='steelblue', alpha=0.85)
    bars2 = ax.bar(x + width/2, f1s,  width, yerr=f1_errs,  capsize=5,
                   label='Mean Macro-F1', color='darkorange', alpha=0.85)

    # Value labels on bars
    for bar in bars1:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9)
    for bar in bars2:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{bar.get_height():.3f}', ha='center', va='bottom', fontsize=9)

    ax.axhline(0.125, color='grey', linestyle=':', linewidth=1,
               label='Chance level (12.5%)')
    ax.set_xticks(x)
    ax.set_xticklabels(groups, fontsize=11)
    ax.set_ylabel('Score')
    ax.set_ylim(0, min(1.0, max(accs + f1s) + 0.15))
    ax.set_title(
        "Feature Group Comparison — Mean Accuracy & Macro-F1\n"
        f"Phase 1 | Test 2 | Logistic Regression | {CONDITION_LABEL}",
        fontsize=12, fontweight='bold'
    )
    ax.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig("P1_T2_feature_group_comparison.png", dpi=150, bbox_inches='tight')
    plt.show()
    print("Saved: P1_T2_feature_group_comparison.png")


def plot_summary_table(all_summaries):
    """
    Render the results as a clean table figure for easy reference.
    """
    col_labels = ['Feature Group', 'N Features',
                  'Mean Acc', 'Std Acc',
                  'Mean Macro-F1', 'Std Macro-F1']
    table_data = [
        [
            s['feature_group'],
            s['n_features'],
            f"{s['mean_accuracy']:.4f}",
            f"{s['std_accuracy']:.4f}",
            f"{s['mean_macro_f1']:.4f}",
            f"{s['std_macro_f1']:.4f}",
        ]
        for s in all_summaries
    ]

    fig, ax = plt.subplots(figsize=(11, 2.5))
    ax.axis('off')
    tbl = ax.table(
        cellText=table_data,
        colLabels=col_labels,
        cellLoc='center',
        loc='center'
    )
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(10)
    tbl.scale(1.2, 1.6)

    # Header styling
    for j in range(len(col_labels)):
        tbl[0, j].set_facecolor('#2c5f8a')
        tbl[0, j].set_text_props(color='white', fontweight='bold')

    # Highlight best accuracy row
    best_acc_idx = int(np.argmax([s['mean_accuracy'] for s in all_summaries]))
    for j in range(len(col_labels)):
        tbl[best_acc_idx + 1, j].set_facecolor('#d4edff')

    ax.set_title(
        "Phase 1 | Test 2 — Feature Group Results Summary\n"
        f"Logistic Regression | {CONDITION_LABEL}  "
        "(blue highlight = best accuracy)",
        fontsize=10, fontweight='bold', pad=12
    )
    plt.tight_layout()
    plt.savefig("P1_T2_results_table.png", dpi=150, bbox_inches='tight')
    plt.show()
    print("Saved: P1_T2_results_table.png")

# =============================================================================
# MAIN
# =============================================================================

if __name__ == "__main__":
    # 1. Load dataset once — subset columns per group inside run_loocv
    X_full, y, actors = load_dataset(DATASET_PATH, SAMPLE_RATE)

    nan_count = int(np.isnan(X_full).sum())
    inf_count = int(np.isinf(X_full).sum())
    print(f"Feature matrix shape : {X_full.shape}")
    print(f"NaN: {nan_count} | Inf: {inf_count}")
    if nan_count > 0 or inf_count > 0:
        print("WARNING: NaN/Inf present — check feature extraction.")
    else:
        print("Feature matrix is clean.\n")

    # 2. Run LOOCV for each feature group
    print("="*70)
    print(f"PHASE 1 | TEST 2 — Feature Group Importance | {CONDITION_LABEL}")
    print("="*70)

    all_summaries  = []
    all_fold_data  = {}

    for group_name, feature_indices in FEATURE_GROUPS.items():
        summary, fold_metrics, y_true, y_pred = run_loocv(
            X_full, y, actors, feature_indices, group_name
        )
        all_summaries.append(summary)
        all_fold_data[group_name] = {
            'fold_metrics': fold_metrics,
            'y_true':       y_true,
            'y_pred':       y_pred,
        }

    # 3. Print full summary table
    print("\n" + "="*70)
    print("RESULTS SUMMARY")
    print("="*70)
    df_summary = pd.DataFrame(all_summaries)
    print(df_summary.to_string(index=False))

    # 4. Per-class report for each group
    print("\n" + "="*70)
    print("PER-CLASS CLASSIFICATION REPORTS")
    print("="*70)
    for group_name, data in all_fold_data.items():
        print(f"\n--- {group_name} ---")
        print(classification_report(
            data['y_true'], data['y_pred'],
            labels=EMOTION_LABELS,
            target_names=EMOTION_LABELS,
            zero_division=0
        ))

    # 5. Plots
    plot_group_comparison(all_summaries)
    plot_summary_table(all_summaries)

    # 6. Save CSVs
    df_summary.to_csv("P1_T2_summary.csv", index=False)
    print("Saved: P1_T2_summary.csv")

    # Per-fold breakdown for all groups
    all_fold_rows = []
    for group_name, data in all_fold_data.items():
        for m in data['fold_metrics']:
            m['feature_group'] = group_name
            all_fold_rows.append(m)
    pd.DataFrame(all_fold_rows).to_csv("P1_T2_fold_metrics.csv", index=False)
    print("Saved: P1_T2_fold_metrics.csv")

    print("\nDone.")
```
